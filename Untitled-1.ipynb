```python
#Работа со списком

#создание списка
group_list = ['Natasha', 'Liza', 'Maksim']
another_group_list = list ()

print(group_list)
print(another_group_list)

#преобразование других типов данных в списки 
spisok = ('Iliya', 'Sasha')
another_group_list = list(spisok)
print(another_group_list)


#получение элемента с помощью конструкции
print(group_list[1])
print(group_list[-1])

#список списков
all_group_list = [group_list, 'Pasha', another_group_list]
print(all_group_list)

#изменение элемента с помощью конструкции
all_group_list[0]='Konstantin'
print(all_group_list)

#извлечение элементов с помощью диапазона смещений
print(all_group_list[0:2])
print(all_group_list[::2])#вывод нечетного элемента
print(all_group_list[::-1]) #инверсия списка

#добавление элемента в конец списка
all_group_list.append('Anna')
print(all_group_list)

#объединение списков
group_list.extend(another_group_list)
print(group_list)
group_list +=all_group_list
print(group_list)

#добавление элемента с помощью функции insert()
all_group_list.insert(0, 'Lika')
print(all_group_list)

all_group_list.insert(3, 'Dasha')
print(all_group_list)

#удаление элемента
del all_group_list[0]
print(all_group_list)

all_group_list.remove('Pasha')
print(all_group_list)

all_group_list.pop(0)
print(all_group_list)

#определение смещения элемента по значению с помощью функции index()
print(all_group_list.index('Dasha'))

# проверка на наличие элемента в списке 
print('Dasha' in all_group_list)   #вернет true
print('Dasha' in another_group_list)  # вернет false, так как не входит 

all_group_list.append('Dasha')

#определение количества включений значения

print(all_group_list.count('Dasha'))  #2
print(all_group_list.count('Anna'))   #1
print(all_group_list.count('A'))   #0

#изменение порядка элементов
another_group_list.sort() #сортировка самого списка
print(another_group_list)

sorted_spisok = sorted(spisok)  # отсортированная копия списка
print(sorted_spisok)

#получение длины списка
print(len(all_group_list))


#копирование списков
a=[1,2,3]
b=a.copy()
print(b)

a[0]=50
print(a)
print(b)
```

```python
#словари
#создание 
empty_dict = {}
new_dict = {
    "day1" : "Понедельник",
    "day2" : "Вторник",
    "day3" : "Среда",
    "day4" : "Четверг",
    "day5" : "Пятница",
    "day6" : "Суббота",
    "day7" : "Воскресенье",
}
print(new_dict)


#добавление или изменение элемента
new_dict['day1'] = 'Monday'
print(new_dict)

#объединение словарей
others_dict = {'1':'1', '2':'2', '3':'3'}
new_dict.update(others_dict)
print(new_dict)

#удаление элементов по ключу 
del new_dict['1']
print(new_dict)

#удаление всех элементов
others_dict.clear()
print(others_dict)
#проверка на наличие элемента
print('2' in new_dict)
print('8' in new_dict)

#получение элемента
print(new_dict['day1'])

```

```python
#множества 
#создание 
empty_set = set()
empty_new_set = {1, 2, 3,4 }
print(empty_new_set)

#пересечение

a= {3,4,5}
print(a & empty_new_set)

#объединение
print(a.union(empty_new_set))
#разность множеств
print(a.difference(empty_new_set))

#исключающее или 
print(a.symmetric_difference(empty_new_set))

#является ли множество подмножеством другого
print(a.issubset(empty_new_set))
print(a.issubset(a))
```
